```python
df.head()
```

# Create Feature Matrix and Target Vector

```python
feature_matrix = df.drop(columns = ['next_pitch'])
target_vector = df['next_pitch']

feature_matrix.shape, target_vector.shape
```

# Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(feature_matrix, target_vector)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Instantiate Model

```python
#Loigistic model
clf1 = Pipeline(steps = [('scaler', RobustScaler()),
                         ('logistic', LogisticRegression(
                             solver = 'lbfgs',
                             multi_class = 'multinomial',
                             random_state = 1))])

# Tree Model
clf2 = Pipeline(steps = [('scaler', RobustScaler()),
                         ('forest', RandomForestClassifier(
                             random_state = 1))])

# Baysian Model
clf3 = Pipeline(steps = [('scaler', RobustScaler()),
                         ('bayes', GaussianNB())])

# Support Vector Model
clf4 = Pipeline(steps = [('scaler', RobustScaler()),
                         ('svm', LinearSVC(
                            penalty = 'l2',
                            loss = 'squared_hinge',
                            dual = False,
                            multi_class = 'ovr',
                            fit_intercept = True,
                            tol = .001,
                            C = 2.2,
                            max_iter = 1000
                        ))])


eclf = VotingClassifier(estimators=[
    ('lr', clf1),
    ('rf', clf2),
    ('gnb', clf3),
    ('svm', clf4)
    ],
    voting='hard')

params = {
    'lr__logistic__C': [1, 3, 5],
    'lr__logistic__max_iter': [10000, 20000],
    'rf__forest__n_estimators': [20, 200],
    'rf__forest__max_depth': [20, 200],
    'svm__svm__intercept_scaling': [800, 900, 1000, 1100, 1200],
}


grid = GridSearchCV(
    estimator = eclf,
    scoring = 'accuracy',
    param_grid = params,
    refit = True,
    cv = 10,
    verbose = 10,
    n_jobs = -1)
```

# Train Model

```python
grid.fit(x_train, y_train)
```

```python
model = grid.best_estimator_
model
```

```python
pd.DataFrame(grid.cv_results_).sort_values(by = 'mean_test_score').head()
```

# Evaluate Model

```python
accuracy_score(model.predict(x_train), y_train)
```
